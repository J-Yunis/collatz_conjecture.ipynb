```python
import matplotlib.pyplot as plt

collatz_numbers = []
a = int(input("number for collatz conjecture"))
while a != 1:
    print(a)
    collatz_numbers.append(a)
    if a %2 == 0:
           a = a / 2 
    else:
       a = a * 3 +1

collatz_numbers.append(a)
x_values = list(range(1, len(collatz_numbers) + 1))
plt.plot(x_values, collatz_numbers)
plt.xlabel('Step number')
plt.ylabel('Collatz number')
plt.show
```
